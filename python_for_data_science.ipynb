## Creating Data Frames

```python
# To make data frames, the most usual way is a dictionary with lists associated to each column:
number = [1,2,3,4]
names = ['Danilo','Eduardo','Renato','Andrea']
pd.DataFrame({
    'names': names,
    'numbers': number
})
```

```python
# The other way is with array of dictionaries
pd.DataFrame([
    {'name':'Danilo', 'numbers':1},
    {'name':'Eduardo', 'numbers':2},
    {'name':'Renato', 'numbers':3},
    {'name':'Andrea', 'numbers':4},
])
```

# Plotting

```python
# Get a summary of a data frame - numerical columns
iris.describe()
```

```python
# Pandas Data Frame have a plot() method for plotting, just have to specific columns.
iris.plot.scatter('sepal_length','sepal_width')
```

```python
# But lets use ggplot that is much better
from ggplot import *

ggplot(aes(x = 'sepal_length', y = 'sepal_width', color = 'species'), data = iris) + \
    geom_point() + \
    theme_bw() + \
    ggtitle('My title')
```

## dplyr (dfply) happiness!!

```python
from dfply import *
```

```python
(iris >>
    select(X.sepal_length, X.species) >>
    mask(X.species == 'setosa') # mask is the same as filter() in R.
)
```

```python
(
iris >>
    group_by(X.species) >>
    summarise(mean_sepal_length = X.sepal_length.mean()) >>
    arrange(X.mean_sepal_length, ascending = True)
)
```
